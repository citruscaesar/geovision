```python
#SpaceNet.extract("sn7", "buildings")

location_code = {
    "acc": "Accra",
    "kam": "Kampala",
    "ptn": "Pointe-Noire",
    "znz": "Zanzibar",
    "gao": "Ngaoundere",
    "mah": "Mahe Island",
    "dar": "Dar Es Salaam",

    "mon": "Mon?",
    "kin": "Kinshasa",
    "nia": "Niamey"
}

def _get_label_name(x: str) -> str | None:
    parts = x.split('/')
    if parts[0] == "test":
        return None
    parts[-1] = parts[-1].replace(".tif", ".geojson")
    parts[-2] = parts[-2] + "-labels"
    return '/'.join(parts)

def _get_location(x: str) -> str | None:
    parts = x.split('/')
    if parts[0] == "test":
        return None
    return location_code[parts[-3]]

def _get_metadata(x: str):
    with rio.open(OpenCities.local / "staging" / x) as raster:
        return (raster.count, raster.height, raster.width, raster.crs.to_epsg(), raster.transform.to_gdal())

df = (
    OpenCities.load("index", "staging", "opencities")
    .assign(image_path = lambda df: df["image_path"].apply(lambda x: '/'.join(x.as_posix().split('/')[-4:]).removeprefix("staging/")))
    .assign(label_path = lambda df: df["image_path"].apply(_get_label_name))
    .assign(location = lambda df: df["image_path"].apply(_get_location))
    .assign(split = lambda df: df["image_path"].str.split('/', expand=True)[0])
)
df[["num_bands", "image_height", "image_width", "crs", "transform"]] = pd.DataFrame(df["image_path"].apply(_get_metadata).to_list(), index = df.index)
```
